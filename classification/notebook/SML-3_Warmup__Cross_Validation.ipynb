#####Hint:

A parameter grid is a dictionary. The keys say where the parameter is in the pipeline while the values provide a list or range of options for the algorithm to try out during the cross validation search. <br><br>
The code for this looks something like the following:
```
param_grid = {
  'transformer__parameter': [option1, option2],
  'model__parameter': range(lowest, highest)
}
```
You would replace "transformer" with a specific transformer like `simpleimputer`,  "parameter" with specific parameters like `max_depth`, and so on. In each case, we have some part of the pipeline, then two underscores __, and then the name of parameter we want to try different values of.
